```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
```

```python
iris
```

```python
y = iris.target
x = iris.data
```

```python
x.shape
```

```python
y
```

```python
x_dr = PCA(2).fit_transform(x)
```

```python
x_dr.shape
```

```python
x_dr[y == 0,0],x_dr[y == 0,1] #for first type (y=0), value of x1 and x2
```

```python
plt.figure()
plt.scatter(x_dr[y == 0,0],x_dr[y == 0,1],c='red',label=iris.target_names[0])
plt.scatter(x_dr[y == 1,0],x_dr[y == 1,1],c='blue',label=iris.target_names[1])
plt.scatter(x_dr[y == 2,0],x_dr[y == 2,1],c='green',label=iris.target_names[2])
plt.legend()
plt.show()
```

```python
pca = PCA(2)
pca = pca.fit(x)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
pca_line = PCA().fit(x) #no decrease of dimension but used for plotting to find a suitable n_components according to cumulative variance
pca_line.explained_variance_ratio_
```

```python
np.cumsum(pca_line.explained_variance_ratio_)
```

```python
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4])
plt.xlabel('# of components after dimension reduction')
plt.ylabel('cumulative explaained variance')
plt.plot()
```

```python
# PCA for noise filtring
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
digits.images.shape
```

```python
def plot_digits(data):
    fig,axes = plt.subplots(4,10,figsize=(10,4),subplot_kw={'xticks':[],'yticks':[]})
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8),cmap='binary')
    
```

```python
plot_digits(digits.data)
```

```python
#artificial noise
np.random.RandomState(57)
noise = np.random.normal(digits.data,2)
plot_digits(noise)
```

```python
pca = PCA(0.5,svd_solver = 'full').fit(noise)
x_dr = pca.transform(noise)
x_dr.shape
```

```python
without_noise = pca.inverse_transform(x_dr)
without_noise.shape
```

```python
plot_digits(without_noise)
```
